```python
 import pandas as pd
```

```python
df = pd.read_csv('enriched_dataset.csv')
```

```python
df = df.rename(columns={'Price (in rupees)': 'Price(in dollars)'})
```

```python
df.head()
```

```python
df['Price(in dollars)'] = df['Price(in dollars)'].astype(str).replace('nan', '0.0').astype(float)
```

```python
df['overlooking'] = df['overlooking'].astype(str).replace('nan', 'Not specified')
df['Car Parking'] = df['Car Parking'].astype(str).replace('nan', 'Not specified')
df['Ownership'] = df['Ownership'].astype(str).replace('nan', 'Not specified')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Status'] = df['Status'].astype(str).replace('nan', 'Not specified')
df['Transaction'] = df['Transaction'].astype(str).replace('nan', 'Not specified')
df['Furnishing'] = df['Furnishing'].astype(str).replace('nan', 'Not specified')
```

```python
df.info()
```

```python
df.to_csv('Cleanedenricheddata.csv', index=False)
```
